```python
class Node:
  def __init__(self, e, n):
    self.elem = e
    self.next = n
```

```python
class arrayStack:
  def __init__(self, s):
    self.stackArr = [None]*len(s)
    self.top = -1
  def push(self, elem):
    if self.top == len(self.stackArr)-1:
      return "Overflowed"
    self.top += 1
    self.stackArr[self.top] = elem
  def pop(self):
    if self.isEmpty():
      return "Underflowed"
    temp = self.stackArr[self.top]
    self.stackArr[self.top] = None
    self.top -= 1
    return temp
  def peek(self):
    if self.isEmpty():
      return "Underflowed"
    return self.stackArr[self.top]
  def isEmpty(self):
    if self.top == -1:
      return True
    else:
      return False
```

```python
class Stack:
  def __init__(self):
    self.top = None
  def push(self, elem):
    n = Node(elem, self.top)
    self.top = n
  def pop(self):
    if self.isEmpty():
      return "Underflowed"
    temp = self.top.elem
    self.top = self.top.next
    return temp
  def peek(self):
    if self.isEmpty():
      return "Underflowed"
    return self.top.elem
  def isEmpty(self):
    if self.top == None:
      return True
    else:
      return False
```

```python
def removeStar(s):
  store = Stack()
  for i in range(len(s)):
    if s[i] == '*':
      store.pop()
    else:
      store.push(s[i])
  ans = ""
  while not store.isEmpty():
    ans = store.pop() + ans
  return ans

print(removeStar("ace***"))
```
