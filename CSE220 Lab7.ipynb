```python
class KeyIndex:
    def __init__(self,a):
        m = a[0]
        n = a[0]
        for i in range(1, len(a)):
          if a[i] > m:
            m = a[i]
        for i in range(1, len(a)):
          if a[i] < n:
            n = a[i]
        if n>=0:
            self.s = 0
            self.aux = [0] *(m+1)
            for i in range(len(a)):
                self.aux[a[i]] +=1
        else:
            self.s = 0-n
            self.aux = [0] * (m+1+(0-n))
            for i in range(len(a)):
                self.aux[a[i] +(0-n)] +=1
    
    def search(self,elem):
        if elem < self.aux[0]:
            return False
        if (elem+self.s) > len(self.aux):
            return False
        if self.aux[elem+self.s] ==0:
            return False
        else:
            return True
    
    def sort(self):
        count,index = 0,0
        for ele in range(len(self.aux)):
            count+=self.aux[ele]
        a = [0] *count
        for i in range(len(self.aux)):
            if self.aux[i] !=0:
                for j in range(self.aux[i]):
                    a[index] = i -self.s
                    index+=1
        return a
    
a = [10,20,-30,50,-30,10,-22]
myObj = KeyIndex(a)
print(myObj.search(15))
print(myObj.search(20))
print(myObj.search(-40))
print(myObj.search(60))
print(myObj.sort())
```
