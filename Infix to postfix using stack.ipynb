```python
class Node:
  def __init__(self, e, n):
    self.elem = e
    self.next = n
```

```python
class Stack:
  def __init__(self):
    self.top = None

  def push(self, obj):
    n = Node(obj, self.top)
    self.top = n

  def pop(self):
    temp = self.top.elem
    self.top = self.top.next
    return temp
  
  def peek(self):
    return self.top.elem
  
  def isEmpty(self):
    if self.top == None:
      return True
    else:
      return False
```

```python
def balancedParantheses(s):
  store = Stack()
  for i in range(len(s)):
    if s[i] == '(':
      store.push(s[i])
    elif s[i] == '{':
      store.push(s[i])
    elif s[i] == '[':
      store.push(s[i])
    elif s[i] == ')':
      if store.isEmpty() and store.peek() != '(':
        return False
      else:
        store.pop()
    elif s[i] == '}':
      if store.isEmpty() and store.peek() != '{':
        return False
      else:
        store.pop()
    elif s[i] == ']':
      if store.isEmpty() and store.peek() != '[':
        return False
      else:
        store.pop()
  return store.isEmpty()
```

```python
def infixToPostfix(expression):
    Operators = ['+', '-', '*', '/', '%', '(', ')', '^', '{', '}', '[', ']']
    Priority = {'+':1, '-':1, '%': 2, '*':3, '/':3,'//': 3, '^':4, '**': 4} 
    store = Stack()
    output = ""
    i = 0
    while i < len(expression):
      if expression[i] not in Operators:
        output += expression[i]
      elif expression[i] == '(':
        output += " "
        store.push(expression[i])
      elif expression[i] == '{':
        output += " "
        store.push(expression[i])
      elif expression[i] == '[':
        output += " "
        store.push(expression[i])
      elif expression[i] == ')':
        output += " "
        while store.peek() != '(':
          output += store.pop() + " "
        output += " "
        store.pop()
      elif expression[i] == '}':
        output += " "
        while store.peek() != '{':
          output += store.pop() + " "
        output += " "
        store.pop()
      elif expression[i] == ']':
        output += " "
        while store.peek() != '[':
          output += store.pop() + " "
        output += " " 
        store.pop()
      else:
        output += " "
        while (not store.isEmpty()) and store.peek() != '(' and store.peek() != '{' and store.peek() != '[' and Priority[expression[i]] <= Priority[store.peek()]:
          output += store.pop() + " "
        output += " "
        if i <= len(expression)-2 and expression[i] == '/' and expression[i+1] == '/':
          store.push(expression[i]+expression[i+1])
          i += 1
        elif i <= len(expression)-2 and expression[i] == '*' and expression[i+1] == '*':
          store.push(expression[i]+expression[i+1])
          i += 1
        else:
          store.push(expression[i])
      i += 1
    output += " "
    while not store.isEmpty():
      output += store.pop() + " "
    return output
expression = "180+3792-29*3+(5^2)*329/89/2*5+967"
print(balancedParantheses(expression))
print('infix notation:',expression)
print('postfix notation:',infixToPostfix(expression))
temp = infixToPostfix(expression)
temp = temp.split()
real = "KL+MN*-OP^W*U/V/T*+Q+"
t = "K+L-M*N+(O^P)*W/U/V*T+Q"
s = infixToPostfix(t)
print('infix notation:',t)
print('postfix notation:',infixToPostfix(t))
```
