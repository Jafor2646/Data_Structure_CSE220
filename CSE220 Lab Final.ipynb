```python
class Node:
  def __init__(self, e ,n = None):
    self.elem = e
    self.next = n
class stack:
  def __init__(self):
    self.top = None
  def push(self, v):
    n = Node(v, self.top)
    self.top = n
  def pop(self):
    if self.top == None:
      return "stack Underflowed"
    temp = self.top
    self.top = self.top.next
    return temp.elem
  def peek(self):
    if self.top == None:
      return "stack Underflowed"
    return self.top.elem
  def isEmpty(self):
    if self.top == None:
      return True
    return False
n = 5
given = [5,1,2,4,3]
main = [None]*n
index = 0
store = stack()
for i in range(len(given)-1):
  if given[i] <= given[i+1]:
    main[index] = given[i]
    index += 1
  else:
    store.push(given[i])
main[index] = given[n-1]
index += 1
while not store.isEmpty():
  main[index] = store.pop()
  index+= 1
flag = True
for i in range(len(main)-1):
  if main[i] > main[i+1]:
    flag = False
    break
if(flag):
  print("YES")
else:
  print("NO")
```
