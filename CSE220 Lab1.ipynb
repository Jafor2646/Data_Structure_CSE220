```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
  
  for i in range(k):
    j = 0
    while j < len(source)-1:
      source[j] = source[j+1]
      j+=1
    source[len(source)-1] = 0
  return source



# Test 02: Rotate Left k cell
def rotate_left(source, k):
  # TO DO
  for i in range(k):
    temp = source[0]
    j = 0
    while j < len(source)-1:
      source[j] = source[j+1]
      j += 1
    source[len(source)-1] = temp
  # Hint, You can write a function for left rotate once and then use it
  return source


# Test 03: Shift Right k cell
def shift_right(source, k):
  # TO DO
  for i in range(k):
    j = len(source)-1
    while j > 0:
      source[j] = source[j-1]
      j -= 1
    source[0] = 0
  # Hint, You can write a function for right shift once and then use it
  return source



# Test 04: Rotate Right k cell
def rotate_right(source, k):
  # TO DO
  for i in range(k):
    temp = source[len(source)-1]
    j = len(source)-1
    while j > 0:
      source[j] = source[j-1]
      j -= 1
    source[0] = temp
  # Hint, You can write a function for right rotate once and then use it
  return source



# Test 05: Remove an element from an array
def remove(source, idx):
  # TO DO
  j = idx+1
  while j < len(source):
    source[j-1] = source[j]
    j+=1
  source[len(source)-1] = 0

  return source



# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
  # TO DO
  i = 0
  while i < len(source):
    if source[i] == element:
      source = remove(source, i)
    else:
      i += 1
  return source



# Test 07: Splitting an Array
def split_array(a):
  # TO DO
    available=False
    if len(a)<2:
        return False
    else:
        for i in range(1,len(a)):
            l_sum=0
            r_sum=0
            for j in range(i):
                l_sum += a[j]
            for j in range(i,len(a)):
                r_sum+=a[j]
            if l_sum==r_sum:
                available = True
                break
        return available


# Test 08: Max Bunch Count
def max_bunch(a):
  # TO DO
  max,current =0,0
  for i in range(len(a)):
    if i==0:
      temp = a[i]
      current+=1
    else:
      if a[i] ==temp:
        current+=1
      else:
        temp = a[i]
        current=1
      if current>max:
        max=current
  return max
```
